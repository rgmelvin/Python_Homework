```python
# import required libraries and modules
from pathlib import Path
import csv

#PART I of the assignment was to read in data from two csv files.
#set the path for the menu csv file:
menupath = Path('menu_data.csv')

#initialize line number
line_num = 0

#Initialized a 'menu' list object to hold the contents of menu_data.csv. Did the same for the sales file.
menu =[]

#For reading in the menu data
with open(menupath, 'r') as menufile:
    menureader = csv.reader(menufile, delimiter=',')

    #I am telling the software there is a header and I am skipping it
    header = next(menureader)
    line_num += 1

    #I am initializing lists for each column of the menu csv file
    item_ls = []
    category_ls = []
    description_ls = []
    price_ls =[]
    cost_ls = []

    #Iterate over rows of the menu csv file:
    for row in menureader:
        #set variables for the items in each row (column variables):
        item = row[0]
        category = row[1]
        description = row[2]
        price = float(row[3])
        cost = float(row[4])

        #Creating lists by appending items to form columns
        item_ls.append(item)
        category_ls.append(category)
        description_ls.append(description)
        price_ls.append(price)
        cost_ls.append(cost)

    #Creating a list of lists with assigned variables for the menu list (initiated above)
    menu.append(item_ls)
    menu.append(category_ls)
    menu.append(description_ls)
    menu.append(price_ls)
    menu.append(cost_ls)

#Set path for the sales csv file:
salespath = Path('sales_data.csv')

#Initialize sales list:
sales = []

#Read in the sales data:
with open (salespath, 'r') as salesfile:
    salesreader = csv.reader(salesfile, delimiter=',')

    #Skipping the header
    header = next(salesreader)
    line_num += 1

    #Initializing lists for each column of the sales csv file
    line_item_id_ls = []
    date_ls = []
    credit_card_number_ls = []
    quantity_ls = []
    menu_item_ls = []

    #Iterating over rows of the sales csv file:
    for row in salesreader:
        #setting variables for hte items in each row (col variables):
        line_item = row[0]
        date = row[1]
        credit_card_number = row[2]
        quantity = int(row[3])
        menu_item = row[4]

        #Creating lists by appending items to form columns
        line_item_id_ls.append(line_item)
        date_ls.append(date)
        credit_card_number_ls.append(credit_card_number)
        quantity_ls.append(quantity)
        menu_item_ls.append(menu_item)

    #Creating list of lists for the sales list (initiated above)
    sales.append(line_item_id_ls)
    sales.append(date_ls)
    sales.append(credit_card_number_ls)
    sales.append(quantity_ls)
    sales.append(menu_item_ls)


#PART II of the assignment was to manipulate the data.
#Making an empty dictionary for the results report:
report = {}

#Initialize the row counter variable
row_count = 0

#Using iteration to pull out unique menu items that had been sold and adding them to the report dictionary.
for menu_item in menu_item_ls:
    if menu_item not in report:
        report[menu_item] = {'01-count': 0, '02-revenue': 0, '03-cogs': 0, '04-profit': 0}


#Using iteration over menu items that have been sold to compile the metrics.
#I calculate the values for each metric and have it added to the report dictionary.
#This is nested and uses the zip function.
for item, price, cost in zip(item_ls, price_ls, cost_ls):
    for menu_item, quantity in zip(menu_item_ls, quantity_ls):
        if item == menu_item:
            report[menu_item]['01-cout'] += quantity
            report[menu_item]['02-revenue'] += price * quantity
            report[menu_item]['03-cogs'] += cost * quantity
            report[menu_item]['04-profit'] += (price - cost) * quantity
        else:
            print(f'{menu_item} does not equal {item}! NO MATCH')

#Writing the contents of the report to a txt file:
lines = report

with open('ramen_report.txt', 'w') as f:
    for key, value in lines.items():
        f.write('%s:%s\n' % (key, value))



```
