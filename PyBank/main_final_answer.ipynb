```python
#import the required modules and libraries
from pathlib import Path
import csv

#set the file path
csvpath = Path('budget_data.csv')

#initialize line number (the purpose of this was not explained to me)
line_num = 0

#open the file. i am opening the file specified by csvpath for the purpose of reading the file. I then call this file 'budgetfile'.
#following that I read the file into 'budgetreader' using csv.reader. I specify ',' as the field delimiter.
with open(csvpath, 'r') as budgetfile:
    budgetreader = csv.reader(budgetfile, delimiter = ',')

    #Skipping the header row
    next(budgetreader)
    line_num =+ 1

    #Initialize list
    mo_ls = []
    pl_ls = []

    #Iterating over the csv file:
    for row in budgetreader:

        #Initialize lists
        mo = row[0]
        pl = int(row[1])
    
        #Create lists from columns
        mo_ls.append(mo)
        pl_ls.append(pl)

#calculate the number of monthly profit/loss periods included
months = len(mo_ls)

#calculate the net total profit/ loss over the  entire period
net = sum(pl_ls)

#initiate list for the change in profit/ loss
delta = []

#iterate to fill delta list
for i in range(1, len(pl_ls)):
    delta.append(pl_ls[i] - pl_ls[i-1])

#Calculations for the report:
ave_chg = round(sum(delta) / len(delta), 2)
max_inc = max(delta)
min_dec = min(delta)

#Get date of max_inc and min_dec
date_max_inc = mo_ls[delta.index(max_inc)]
date_min_dec = mo_ls[delta.index(min_dec)]

#Printing each of these as a check along the way. They are hashed out now
#print(months)
#print(f'{net:,}')
#print(delta)
#print(f'{ave_chg:,}')
#print(f'{max_inc:,}')
#print(date_max_inc)
#print(f'{min_dec:,}')
#print(date_min_dec)

#print the Financial Analysis statement
lines = [f'\n\
    Financial Analysis\n\
    -------------------------------------------------------\n\
    Total Months:                   {months}\n\
    Total:                          $ {net:,}\n\
    Average Change:                 $ {ave_chg:,}\n\
    Greatest Increase in Profits:   {date_max_inc} ($  {max_inc:,})\n\
    Greatest Decrease in Profits:   {date_min_dec} ($ {min_dec:,})']
with open('PyBank_FA_final_ans.txt', 'w') as file:
    for line in lines:
        file.write(line)
```
